```python
from src.promptor import *
from src.constants import *
pd.set_option('display.max_colwidth', None)
```

```python
engine = "text-davinci-003"
max_tokens = 1000
frequency_penalty = 0
presence_penalty = 0
```

```python
temperatury = [0, 0.1, 0.2, 0.3, 0.4, 0.50, 0.60, 0.7, 0.80, 0.90, 1]
```

```python
df_temperatury = []
for t in temperatury:
    df_t = promptor("Následuje složitá věta.", "Složitá věta", "Homo economicus poptával svět, ve kterém by změny vnesené mezi nás těžbou uhlí, růstem měst, tovární výrobou či chemickým průmyslem neměly v představě veřejnosti žádný vliv na životní prostředí: dějiny kapitalismu jsou především dějinami vědecké produkce ignorance dopadu industrializace na náš pozemský svět.", "Zjednoduš citovanou větu:", None, t, frequency_penalty, presence_penalty, max_tokens, engine)
    df_temperatury.append(df_t)
df_temperatury2 = pd.DataFrame()
for x in df_temperatury:
    d_x = frejm(x)
    df_temperatury2 = pd.concat([df_temperatury2, d_x], axis=0, ignore_index=True)
df_temperatury2
```

```python
df_temperatury2.to_json("01 test temperatur.json", orient="records")
```
